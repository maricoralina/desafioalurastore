#1. Análise do faturamento


```python
import pandas as pd

# Carregar os dados
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"
]

lojas = [pd.read_csv(url) for url in urls]

# Calcular faturamento total por loja
faturamentos = [df['Preço'].sum() for df in lojas]

# Apresentar resultados
for i, fat in enumerate(faturamentos, 1):
    print(f'Loja {i}: R$ {fat:,.2f}')
```

# 2. Vendas por Categoria


```python
import pandas as pd

# URLs dos dados das lojas
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"
]

# Carregar os dados das lojas
lojas = [pd.read_csv(url) for url in urls]

# Função para calcular faturamento por categoria
def faturamento_por_categoria(df):
    return df.groupby('Categoria do Produto')['Preço'].sum().sort_values(ascending=False)

# Calcular faturamento por categoria para cada loja
faturamentos_categorias = [faturamento_por_categoria(loja) for loja in lojas]

# Exibir resultados
for i, faturamento in enumerate(faturamentos_categorias, 1):
    print(f"Loja {i} - Faturamento por Categoria:")
    print(faturamento)
    print()
```

# 3. Média de Avaliação das Lojas

```python
import pandas as pd

# URLs dos dados das lojas
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"
]

# Carregar os dados das lojas
lojas = [pd.read_csv(url) for url in urls]

# Função para calcular a média de avaliação das lojas
def media_avaliacao(df):
    return df['Avaliação da compra'].mean()

# Calcular a média de avaliação para cada loja
medias_avaliacao = [media_avaliacao(loja) for loja in lojas]

# Exibir resultados
for i, media in enumerate(medias_avaliacao, 1):
    print(f'Média de Avaliação da Loja {i}: {media:.2f}')
```

# 4. Produtos Mais e Menos Vendidos

```python
import pandas as pd

# URLs dos dados das lojas
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"
]

# Carregar os dados das lojas
lojas = [pd.read_csv(url) for url in urls]

def produtos_mais_menos_vendidos(df):
    contagem = df['Produto'].value_counts()
    return (contagem.idxmax(), contagem.max()), (contagem.idxmin(), contagem.min())

# Calcular para cada loja
resultados = [produtos_mais_menos_vendidos(loja) for loja in lojas]

# Exibir resultados formatados
for i, (mais, menos) in enumerate(resultados, 1):
    print(f"Loja {i}:")
    print(f"  Produto mais vendido: {mais[0]} ({mais[1]} unidades)")
    print(f"  Produto menos vendido: {menos[0]} ({menos[1]} unidades)\n")
```

# 5. Frete Médio por Loja


```python
import pandas as pd

# URLs dos dados das lojas
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"
]

# Carregar os dados das lojas
lojas = [pd.read_csv(url) for url in urls]

# Função para calcular o frete médio por loja
def frete_medio(df):
    return df['Frete'].mean()

# Calcular o frete médio para cada loja
fretes_medios = [frete_medio(loja) for loja in lojas]

# Exibir resultados
for i, frete in enumerate(fretes_medios, 1):
    print(f'Frete médio da Loja {i}: R$ {frete:.2f}')
```

# Relatório

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# URLs dos dados das lojas
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"
]

# Carregar os dados das lojas
lojas = [pd.read_csv(url) for url in urls]
nomes_lojas = [f'Loja {i+1}' for i in range(4)]

# Calcular métricas para cada loja
faturamentos = [df['Preço'].sum() for df in lojas]
medias_avaliacao = [df['Avaliação da compra'].mean() for df in lojas]
fretes_medios = [df['Frete'].mean() for df in lojas]

# Análise de categorias por loja
faturamento_categorias = []
for loja in lojas:
    faturamento_categorias.append(loja.groupby('Categoria do Produto')['Preço'].sum().sort_values(ascending=False))

# Produtos mais e menos vendidos
produtos_mais_vendidos = []
produtos_menos_vendidos = []
for loja in lojas:
    contagem = loja['Produto'].value_counts()
    produtos_mais_vendidos.append((contagem.idxmax(), contagem.max()))
    produtos_menos_vendidos.append((contagem.idxmin(), contagem.min()))

# Configuração de estilo
plt.style.use('seaborn-v0_8')
plt.figure(figsize=(18, 15))

# 1. Gráfico de barras - Faturamento total
plt.subplot(2, 2, 1)
bars = plt.bar(nomes_lojas, faturamentos, color='royalblue')
plt.title('Faturamento Total por Loja', fontsize=14)
plt.ylabel('Faturamento (R$)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adicionar valores nos topos das barras
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'R$ {height:,.0f}',
             ha='center', va='bottom', fontsize=9)

# 2. Gráfico de pizza - Distribuição do faturamento por categorias da Loja 1
plt.subplot(2, 2, 2)
top_categorias = faturamento_categorias[0].head(5)
outros = sum(faturamento_categorias[0].iloc[5:])
dados_pizza = list(top_categorias) + [outros]
labels = list(top_categorias.index) + ['Outros']

plt.pie(dados_pizza, labels=labels, autopct='%1.1f%%', startangle=90,
        colors=sns.color_palette('pastel', len(dados_pizza)),
        wedgeprops={'edgecolor': 'w'})
plt.title('Distribuição de Vendas por Categoria - Loja 1', fontsize=14)

# 3. Gráfico de linhas - Comparativo de indicadores por loja
plt.subplot(2, 1, 2)

# Normalizando os valores para uma escala comparável
norm_faturamento = [fat/max(faturamentos) for fat in faturamentos]
norm_avaliacao = [av/max(medias_avaliacao) for av in medias_avaliacao]
norm_frete_invertido = [min(fretes_medios)/frete for frete in fretes_medios]  # Invertido (menor é melhor)

x = np.arange(len(nomes_lojas))
plt.plot(x, norm_faturamento, 'o-', linewidth=2, label='Faturamento (normalizado)')
plt.plot(x, norm_avaliacao, 's-', linewidth=2, label='Avaliação (normalizada)')
plt.plot(x, norm_frete_invertido, '^-', linewidth=2, label='Eficiência Frete (normalizada)')

# Calculando e plotando o score final (ponderado)
peso_faturamento = 0.5  # 50% importância
peso_avaliacao = 0.3    # 30% importância
peso_frete = 0.2        # 20% importância

scores = []
for i in range(4):
    score = (norm_faturamento[i] * peso_faturamento) + \
            (norm_avaliacao[i] * peso_avaliacao) + \
            (norm_frete_invertido[i] * peso_frete)
    scores.append(score)

plt.plot(x, scores, 'D-', linewidth=3, color='purple', label='Score Final')

plt.title('Comparativo de Indicadores de Desempenho', fontsize=14)
plt.xlabel('Lojas', fontsize=12)
plt.ylabel('Valor Normalizado (0-1)', fontsize=12)
plt.xticks(x, nomes_lojas)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

plt.tight_layout()
plt.savefig('relatorio_vendas.png', dpi=300)
plt.show()

# Determinar qual loja vender (menor score final)
loja_a_vender_idx = scores.index(min(scores))
loja_a_vender = nomes_lojas[loja_a_vender_idx]

# Relatório final
print("\n" + "="*50)
print("RELATÓRIO DE ANÁLISE PARA DECISÃO DE VENDA")
print("="*50)

print("\n1. ANÁLISE DE FATURAMENTO")
for i, fat in enumerate(faturamentos):
    print(f"{nomes_lojas[i]}: R$ {fat:,.2f}")

print("\n2. AVALIAÇÃO MÉDIA DOS CLIENTES")
for i, aval in enumerate(medias_avaliacao):
    print(f"{nomes_lojas[i]}: {aval:.2f} / 5.00")

print("\n3. FRETE MÉDIO")
for i, frete in enumerate(fretes_medios):
    print(f"{nomes_lojas[i]}: R$ {frete:.2f}")

print("\n4. PRODUTOS MAIS VENDIDOS")
for i, produto in enumerate(produtos_mais_vendidos):
    print(f"{nomes_lojas[i]}: {produto[0]} ({produto[1]} unidades)")

print("\n5. PRODUTOS MENOS VENDIDOS")
for i, produto in enumerate(produtos_menos_vendidos):
    print(f"{nomes_lojas[i]}: {produto[0]} ({produto[1]} unidades)")

print("\n6. TOP 3 CATEGORIAS POR FATURAMENTO")
for i, fat_cat in enumerate(faturamento_categorias):
    top3 = fat_cat.head(3)
    print(f"{nomes_lojas[i]}: {', '.join([f'{cat} (R$ {valor:,.2f})' for cat, valor in top3.items()])}")

print("\n" + "="*50)
print("CONCLUSÃO E RECOMENDAÇÃO")
print("="*50)
print(f"\nCom base na análise realizada, recomendamos que o Sr. João venda a {loja_a_vender}.")

# Justificativa específica para a loja recomendada
if loja_a_vender == "Loja 4":
    print(f"\nJustificativa:")
    print(f"- A Loja 4 apresenta o menor faturamento total (R$ {faturamentos[3]:,.2f}), representando apenas {faturamentos[3]/sum(faturamentos)*100:.1f}% do faturamento total.")
    print(f"- Apesar de ter o menor custo de frete (R$ {fretes_medios[3]:.2f}), esta vantagem não compensa o baixo desempenho em vendas.")
    print(f"- Sua avaliação média ({medias_avaliacao[3]:.2f}) é satisfatória, mas não excepcional o suficiente para justificar sua manutenção diante do baixo faturamento.")
elif loja_a_vender == "Loja 1":
    print(f"\nJustificativa:")
    print(f"- Embora a Loja 1 tenha o maior faturamento (R$ {faturamentos[0]:,.2f}), ela apresenta a pior avaliação de clientes ({medias_avaliacao[0]:.2f}).")
    print(f"- O frete médio é o mais alto entre todas as lojas (R$ {fretes_medios[0]:.2f}), indicando ineficiência logística.")
    print(f"- A combinação de baixa avaliação e alto custo logístico representa um risco futuro para o negócio, mesmo com o atual bom faturamento.")

print("\nEsta análise considerou as seguintes ponderações:")
print(f"- Faturamento: {peso_faturamento*100:.0f}% de importância")
print(f"- Avaliação dos clientes: {peso_avaliacao*100:.0f}% de importância")
print(f"- Eficiência logística (frete): {peso_frete*100:.0f}% de importância")

print("\nRecomendações adicionais:")
print("1. Redirecionar os recursos da venda para melhorar as operações das lojas restantes")
print("2. Aplicar as melhores práticas da loja com maior avaliação nas demais unidades")
print("3. Analisar possibilidades de otimização logística, especialmente nas lojas com frete elevado")
```
